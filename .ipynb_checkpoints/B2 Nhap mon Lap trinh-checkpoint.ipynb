### PIP là gì?

PIP là công cụ giúp cài đặt, cập nhật và xoá bỏ các thư viện, module,... của Python.

Lệnh kiểm tra phiên bản PIP

`pip --version`

### Thư viện Turtle
Turtle là thư viện giúp chúng ta vẽ đồ họa 2D

Thư viện Turtle dùng hệ trục tọa độ Oxy với tọa độ (0,0) nằm ngay giữa khung cửa sổ

Cài đặt thư viện Turtle bằng lệnh:

`pip install turtle`

##### Nguồn tài liệu tham khảo: https://docs.python.org/3/library/turtle.html

### Lệnh thêm thư viện Turtle

`import turtle`

### Lệnh tạo đối tượng Turtle
`t = turtle.Turtle()`

t có thể là tên bất kỳ

Thực hiện thử:

### Các lệnh di chuyển

Lệnh tiến tới

`t.forward(5)` hoặc `t.fd(5)`

Lệnh lùi lại

`t.back(5)` hoặc `t.backward(5)` hoặc `t.bk(5)`

Lệnh rẽ phải:

`t.right(45)` hoặc `t.rt(45)`

Lệnh rẽ trái:

`t.left(45)` hoặc `t.lt(45)`


Thực hành về hình vuông:



```python
import turtle

t = turtle.Turtle()

t.fd(100)
t.right(90)
t.fd(100)
t.right(90)
t.fd(100)
t.right(90)
t.fd(100)
t.right(90)
```

### Lệnh lặp

`for i in range(5):`

Vẽ hình vuông bằng lệnh lặp:

```python
import turtle

t = turtle.Turtle()

for i in range(5):
    t.fd(100)
    t.right(90)
```

Thực hành vẽ 2 hình vuông:



##### Gợi ý: sử dụng lệnh `t.penup()` và `t.pendown()`

```python
import turtle

t = turtle.Turtle()

for i in range(4):
    t.fd(100)
    t.right(90)

t.penup()
t.fd(200)
t.pendown()

for i in range(4):
    t.fd(100)
    t.right(90)
```

### Lệnh đổi màu fill và màu nét vẽ

`t.color(màu nét vẽ, màu fill)`

Ví dụ: `t.color("green","yellow")`

### Lệnh bắt đầu fill màu

`t.begin_fill()`

### Lệnh kết thúc fill màu

`t.end_fill()`


Vẽ hình tam giác đầu với màu như sau:



```python
import turtle

t = turtle.Turtle()

t.color("green","pink")
t.begin_fill()
for i in range(3):
    t.fd(100)
    t.left(120)
t.end_fill()
```

<h3>Bài tập</h3>

<h5>Vẽ hình lục giác</h5>

```python
import turtle

t = turtle.Turtle()

for i in range(6):
    t.fd(100)
    t.left(60)
```

<h5>Vẽ hình ngôi sao</h5>



```python
import turtle

t = turtle.Turtle()

t.color("green","yellow")
t.begin_fill()

for i in range(5):
    t.fd(100)
    t.right(144)

t.end_fill()
```

<h5>Vẽ hình tròn</h5>

Không sử dụng lệnh `circle`

```python
import turtle

t = turtle.Turtle()

for i in range(360):
    t.fd(3)
    t.right(1)
```

Có sử dụng lệnh `circle`

```python
import turtle

t = turtle.Turtle()

t.circle(100)
```
